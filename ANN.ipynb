```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('/Users/aromaatieno/Downloads/Breast_cancer/wisconsin.csv')
#del data['Unnamed: 32']

print(data)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Labeling data in columns 3 and 2 as x and y
X = data.iloc[:,2].values
y = data.iloc[:,1].values

#lr = LinearRegression()
#lr.fit(X.reshape(-1,1), y.reshape(-1,1))


#importing the Labelencoder class which concerts strings to numerical data
from sklearn.preprocessing import LabelEncoder

#creating an instance of Labelencoder class
labelencoder_X_1 = LabelEncoder()

#Using the fit_transorm method of the labelencoder class to transform values assigned to the y variable that are currently in string foramt into numerical format and reassigns them back to the y variable
y = labelencoder_X_1.fit_transform(y)
```

```python
#splitting the data to get training and testing data. testing data will make up 0.1/ 10% of our data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.1, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#standardizing input features in the X_train variable using fit_transform function of standard scaler class
X_train = sc.fit_transform(X_train.reshape(-1,1))
#standardizing input features in the X_test variable using transform function of standard scaler class
X_test = sc.transform(X_test.reshape(-1,1))
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
#initiatializing a linear stack of layers that will make up the neural network
classifier = Sequential()
```

```python

#adding a layer with the specified parameters to the neural network
classifier.add(Dense(units=16,input_shape=(X_train.shape[1],), kernel_initializer='uniform' , activation = 'relu', input_dim = 30))

#classifier.add(Dense(256, input_shape=(X.shape[1],), activation='relu', input_dim=30))

#dropping out 10% of the neurons to prevent overfitting
classifier.add(Dropout(rate = 0.1))
```

```python
#adding a second layer to the neural network
#classifier.add(Dense(units=16, kernel_initializer='uniform' , activation = 'relu'))

classifier.add(Dense(16,kernel_initializer='uniform', activation='relu'))  
#dropping out 10% of the neurons to prevent overfitting
classifier.add(Dropout(rate = 0.1))
```

```python
#adding the output layer
#classifier.add(Dense(units=1, kernel_initializer='uniform' , activation = 'sigmoid'))
classifier.add(Dense(1,kernel_initializer='uniform' , activation = 'sigmoid'))
```

```python
#compiling the output network and ensuring to minimize the loss function using an optimizer
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
import tensorflow as tf
```

```python
classifier.fit(X_train, y_train, batch_size = 100, epochs = 512)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print("Our accuracy is {}%" . format (100 -(((cm[0][0] + cm[1][1])/57) * 100)))
```
